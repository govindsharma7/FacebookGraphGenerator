```python
import facebook
token = 'enter access code'
graph = facebook.GraphAPI(token)
profile = graph.get_object("me")
myid=profile['name']
fbgraph={}
friends = graph.get_connections("me", "friends")
friend_list = [friend['name'] for friend in friends['data']]
friend_ids = [friend['id'] for friend in friends['data']]
fbgraph[myid]=friend_ids

friends_of_friends=[]
for fid in friend_ids:
    friend_list=[]
    temp_profile = graph.get_object(fid)
    myid=temp_profile['name']
    temp_friends = graph.get_connections(fid, "friends")
    friend_list = [friend['id'] for friend in temp_friends['data']]
    fbgraph[fid]=friend_list
    friends_of_friends.append(friend_list)
```

```python
import networkx as nx
G=nx.Graph(fbgraph)
```
